```python
import requests
from serpapi import GoogleSearch
title = input("Enter your search Query: ")
params={
    "q":title,
    "hl":"en",
    "engine":"Google_scholar",
    "api_key":"Try your APi key"
}
search=GoogleSearch(params)
response=search.get_dict()
print(response)
```

```python
Searchresults=response['organic_results']
print(Searchresults)
titles=[]
cited_data_links=[]
cited_data_total_number=[]
cited_data_id=[]
for result in Searchresults:
  print(result['title'])
  titles.append(result['title'])
  for key,rs in result.items():
    if key=="inline_links" :
      print(rs['cited_by']['link'])
      cited_data_links.append(rs["cited_by"]["link"])
      cited_data_total_number.append(rs['cited_by']['total'])
      cited_data_id.append(rs['cited_by']['cites_id'])

cited_data_titles=[]
for i in range(len(cited_data_links)):
        print(f" document : {titles[i]} cited : {cited_data_total_number[i]}")


print(cited_data_id)
```
